# Uniform Distribution

A uniform distribution has constant probability within a given range.
 The uniform Distribution is a continues distribution in which all outcomes in an interval [a,b] are equally likely.

 it is defined by parameters:
    
    * a : Lower Bound
    * b : Upper Bound

Example : modelling the probability of a random point along a line segment.

```python


from scipy.stats import uniform

#parameters
a,b=0,10
x_values=np.linspace(a,b,1000)

# plot of PDf
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(x_values,uniform.pdf(x_values,a,b-a),color='blue')
plt.title("uniform distribution pdf(a=0,b=10)")
plt.xlabel("x")
plt.ylabel("PDF")

# plot CDF
plt.subplot(1,2,2)
plt.plot(x_values,uniform.cdf(x_values,a,b-a),color='green')
plt.title("uniform distribution cdf(a=0,b=10)")
plt.xlabel("x")
plt.ylabel("CDF")

plt.tight_layout()
plt.show()
```

```python
# parameter for discrete uniform distribution
low,high=1,6
x_values=np.arange(low,high+1)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.stem(x_values,randint.pmf(x_values,low,high+1),basefmt="")
plt.title(" discrete uniform distribution pdf(low=1,high=6)")
plt.xlabel("x")
plt.ylabel("PDF")

plt.subplot(1,2,2)
plt.step(x_values,randint.cdf(x_values,low,high+1),color="green")
plt.title(" discrete uniform distribution cdf(low=1,high=6)")
plt.xlabel("x")
plt.ylabel("CDF")

plt.tight_layout()
plt.show()
```

# **Bernoulli Distribution**
**The Bernoulli Distribution modls a single experiment with two possible oucomes: succes(with probability `p`) and failure(with probability `1 - p`)**

**it is discrete distrbution and defined by the paramter:**
 - `p:` Prabibility of success Example: Modling a single coin flip where heads is success and tails is failure.

```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
import numpy as np
# parameters
p=0.5

x_values= np.array([0,1])

plt.figure(figsize=(12,6))

# plot PMF
plt.subplot(1,2,1)
plt.stem(x_values, bernoulli.pmf(x_values, p), basefmt=" ")
plt.title("Bernoulli Distribution PMF (p = 0.5)")
plt.xlabel("x")
plt.ylabel("PMF")
plt.show()
```

# Normal Distribution

 The Normal Distrbution, also known as Gaussian distribtion, is a continuous distribution  defined by its symmetric, bell-shaped curve centered around the mean.
  
- It is definde by paramters:
 - `mu`: Mean of the distrbution
 - `sigma`: Standard deviation, which determines the spread of the distribution


- Special case: The Standard Normal Distribution has `mu= 10` and `sigma= 1`.
 - Example: Modeling heights, test scores, or measurement errors,

```python
from scipy.stats import norm


#  paramters for Normal Distribution
mu =1 #mean
sigma = 3
x= np.linspace(-10,10,100)


plt.figure(figsize=(12,6))

# plot PDF
plt.subplot(1,2,1)
plt.plot(x, norm.pdf(x, mu, sigma),color="blue",label="PDF")
plt.title("Normal Distribution PDF")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.legend()


# CDF
plt.subplot(1,2,2)
plt.plot(x, norm.cdf(x, mu, sigma),color="red",label="CDF")
plt.title("Normal Distribution CDF")
plt.xlabel("x")
plt.ylabel("p(x <= x)")
plt.grid()
plt.legend()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
# parametes for poisson distribution
mu = 3
x= np.arange(0,15)
from scipy.stats import poisson

plt.figure(figsize=(12,6))

# plto PMF
plt.subplot(1,2,1)
plt.bar(x, poisson.pmf(x,mu), color="purple")
plt.title("poisson Distrbutin PMF")
plt.xlabel("X")
plt.ylabel(" P(x,=x)")
plt.legend()




# CDF
plt.subplot(1,2,2)
plt.step(x, poisson.cdf(x,mu),color="blue")
plt.title("poisson Distrbutin CDF")
plt.xlabel("X")
plt.ylabel(" p(x <= x")
plt.legend()
plt.show()
```

# Gamma Distribution

The Gamma DIstribution is a continuous distribution defined for psitive values.
it is often used ot model waiting times and is defined by paramters:
 - `k`= Shape parameter
 - `theta`=  Scale parameter (or `1/ rate`)

 Example: modeling the time until a specified number of events occur, susch as time  until a machine fails

```python
k= 2 # shape parameter
theta = 2 # scale parameter
x = np.linspace(0,20,100)
from scipy.stats import gamma

plt.figure(figsize=(12,6))
# plto PDF
plt.subplot(1,2,1)
plt.plot(x,gamma.pdf(x, k, scale= theta ),color="g",label="PDF")
plt.title("Gamma Distribution PDF")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()


# CDF
plt.subplot(1,2,2)
plt.plot(x, gamma.cdf(x,k, scale=theta),color="purple",label="CDF")
plt.title("Gamma Distribution CDF")
plt.xlabel("x")
plt.ylabel("p(x <= x")
plt.grid(True)
plt.legend()
plt.show()
```

# Beta Distribution

The beta distrbution is a continuous probability distribution defined on the interval [0,1].
it is often used to model probabilities or proportions.
Defined by shape parameters:
 - `a` and `b` : Contol the distributions shpae and are typically greater than 0.



 Example : Modeling random variables that represent probabilites , such as the succes rate in A/B testing.

```python
a,b= 2,5
x = np.linspace(0,1,100)
from scipy.stats import beta

plt.figure(figsize=(12,6))
# plot PDF
plt.subplot(1,2,1)

plt.plot(x,beta.pdf(x,a,b),color="c",label="PDF")
plt.title("Beta Distribution PDF")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()

# plot of CDF
# CDF
plt.subplot(1,2,2)
plt.plot(x, beta.cdf(x,a,b),color="m",label="CDF")
plt.title("Beta Distribution CDF")
plt.xlabel("x")
plt.ylabel("p(x <= x")
plt.grid(True)
plt.legend()
plt.show()
```
