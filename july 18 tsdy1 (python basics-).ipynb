```python
r= int(input("enter the rows "))
for i in range (r):
  if i==0 or i==r-1:
    for j in range(r):
      print("*",end=" ")
  else:
    print("*",end="")
    for k in range(1,2*r-2):
      print(" ",end="")
    print("*",end=" ")
  print()
```

```python

def fn(n):
  if n==0:
    return 1
  else:
    return n*fn(n-1)

a=int(input("number: "))
result=fn(a)
print(a,"factorial=",result)
```

```python
def fn(n):
  if n==0:
    return 0
  else:
    return n+fn(n-1)

sum=int(input("number: "))
result=fn(sum)
print("sum of ",sum,"=",result)
```
