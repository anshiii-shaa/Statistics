```python
# import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create sample data
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'B', 'C', 'B', 'C', 'A', 'C', 'B', 'A', 'C'],
    'Scores': [85, 90, 88, 75, 95, 88, 90, 78, 88, 80, 90, 78, 92, 85, 77],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female']
}
```

```python
df=pd.DataFrame(data)
print(df)
```

freaquncy means the number if times a value appears un the data . a table can quickly show us how many times each value appears.

# Frequency Table for Qualitative Data

**catogory**

```python
# frequency table for "category"
category_freq= df["Category"].value_counts()
print("frequncy table for category:\n",category_freq)
```

```python
# plot for "category"
sns.countplot(x="Category",data=df,palette="viridis")
plt.title("Frequancy distribution: Category")
plt.xlabel("Category")
plt.ylabel("Frequency")
```

```python
# frequency table for "Gender"
gender_freq=df["Gender"].value_counts()
print("frequency table for: 'Gender':\n",gender_freq)
```

```python
# plot for "Gender"
sns.countplot(x="Gender",data=df,palette="magma")
plt.title("Frequancy distribution: Gender")
plt.xlabel("Gender")
plt.ylabel("Frequency")
```

# Frequncy Table for Numerical Data

```python
# frequency table for "scores"
scores_freq=df["Scores"].value_counts()
print("frequency table for: 'scores':\n",scores_freq)
```

```python
# plot for "Gender"
sns.countplot(x="Scores",data=df,palette="rainbow")
plt.title("Frequancy distribution: Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
```

**if the data has many diffrent values,it is easier to use intervals of values to present them in a table.**

```python
# crating bins for "scores"
df["Score_bins"]=pd.cut(df["Scores"],bins=[70,80,90,100],labels=["70-80","80-90","90-100"])

# frequncy table for "Score_bins"
score_bins_freq=df["Score_bins"].value_counts().sort_index()
print("\nFrequency Table for 'Scores_bins':\n",score_bins_freq)
```

```python
# plotting frequency distribution for score

sns.countplot(x="Score_bins",data=df,palette="cool",legend=True,hue="Score_bins")
plt.title("Frequancy distribution: Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```
