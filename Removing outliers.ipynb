```python
import numpy  as  np

```

```python
# sample data
data=[7,15,36,39,40,41,42,43,47,49]
```

**QUARTILE** kand pidikan **quantile** equation!!

```python
# calculate q1, q2 and q3
q1= np.quantile(data, 0.25)
q2= np.quantile(data, 0.50)
q3= np.quantile(data, 0.75)

print(q1,"\n")
print(q2,"\n")
print(q3,"\n")
```

```python
# calculate the IQR
iqr= q3-q1
iqr
```

**BOX PLOT**

```python
import matplotlib.pyplot as plt

# boxplot to visualize quartiles and iqr
plt.boxplot(data,vert=False)
plt.title(" Boxplot of data")
plt.ylabel("value")
plt.show()
```

**OUTLIER**

```python
lower_bound = q1 - 1.5* iqr
upper_bound = q3 + 1.5* iqr

print(lower_bound)
print(upper_bound)
```

the whiskers extend to the

```python
# identify outliers
outliers= [x for x in data if x < lower_bound or x > upper_bound]
# any data point below the lower bound or above the upper bound is considered an outliers.

# output the results
outliers
```
